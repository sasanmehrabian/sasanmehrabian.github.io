### It is hard to write custome queries each time. Lets use the power of Pandas/Python by getting everything into Pandas DF:

```python
# Use "for loops" to read each table in the DB and store it as a df with the same name as the table.
#You should see a set of Pandas df like this :




```

```python
import pandas as pd 
from pandasql import sqldf

# get the names of the tables and store it in an array 
data_names = []
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
for line in cur.fetchall():
    data_names.append(line[0])

# make an array of the names the data frames should be    
data_frame_names = [] 

for j in data_names : 
    data_frame_names.append(j+'_df')
data_frame_names


# get data from the data base 
for x,n in enumerate(data_names) : 

    format_str1 = "SELECT * FROM {name};"
    sql_command1 = format_str1.format(name = n)
    
    cur.execute(sql_command1)
    exec(data_frame_names[x] + '= pd.DataFrame(cur.fetchall())')

# get column name values
col = list(range(len(data_names)))

for x,n in enumerate(data_names) : 
    
    col_names = []
    format_str2 = "PRAGMA table_info({name});"
    sql_command2 = format_str2.format(name = n)
    cur.execute(sql_command2)
    for line in cur.fetchall():
        col_names.append(line[1])
    col[x] = list(col_names)

# assign column names to data frames 

for x,i in enumerate(col) :
    # check to make sure the data_frames are not empty to avoid errors 
    if (eval(data_frame_names[x])).columns.size == 0:
        pass
    else : 
        (eval(data_frame_names[x])).columns = i
```

### Explain why using "Total" is not the best meric ?

This is becasue total will sum up all the different attributes together but does not tell which species is the strongest in each category. It would be more useful to find the species with the largest values in each catergory. It is therefore hard to determine which pokemon is the strongest and the weakest because a pokemon may be weak in some areas but strong in others. It would be easier to compare the pokemon categories with each other. 

```python
# Instead, try a different metric: standardize the six statistic columns independently by converting each 
# value into a z-score so when we do take the sum, we account for the variation in the each statistic using 
# its mean and standard deviation across all Pokemon species. 
```

```python
# create names for z-stats table 
z_names = ['id_z','hp_z','attack_z','defense_z','sp_atk_z','sp_def_z','speed_z','total_z']

# get the z score for every column 

for x,i in enumerate(poke_statistics):
    z = (poke_statistics[i] - np.mean(poke_statistics[i]))/(np.std(poke_statistics[i]))
    exec(z_names[x] + " =z")
    
# create data frame 

z_stats = pd.DataFrame({'hp_z' : hp_z,
                        'attack_z' : attack_z,
                        'defense_z' : defense_z, 
                        'sp_atk_z': sp_atk_z,
                        'sp_def_z' : sp_def_z,
                        'speed_z' : speed_z,
                        'total_z' : total_z
})

z_stats.head()
```

```python
# Define a new column, strength, as the sum of the z-scores of each statistic—the higher this value, 
# the stronger the Pokemon.
```

```python
pokedex['strength'] = z_stats.sum(axis=1)
df.head()
```
